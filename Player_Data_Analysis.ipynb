```python
df_sorted=df.sort_values(by='Overall',ascending=False)
df_sorted
```

```python
df_sorted.head(20)
```

```python
data = df_sorted['Overall'].head(15)
d1=df_sorted['Name'].head(15)
plt.pie(data,labels=d1, autopct='%1.1f%%', startangle=120)
plt.title("Players VS OVR" )
plt.show()
```

Displaying the pie chart comparison between  playes and  their  respective OVR.

```python
data = df['Overall'].head(15)
d1=df['Name'].head(15)
plt.barh(d1, data)
plt.xlabel('Name')
plt.ylabel('Overall')
plt.title("Players VS OVR" )
plt.show()
```

Displaying the bar graph comparison between  playes and  their  respective OVR.

```python
data = df['Overall'].head(13)
d1=df['Name'].head(13)
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])


plt.scatter(data,d1,c=colors)
plt.ylabel('Overall')
plt.title("Players VS OVR" )
plt.grid(True)
plt.show()
```

Displaying the scatter chart comparison between  playes and  their  respective OVR.
